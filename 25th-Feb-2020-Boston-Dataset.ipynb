```python
columns = df.columns
```

```python
plt.figure(figsize=(10,100))
plt.subplots_adjust(hspace =0.5, wspace=0.5)
for i in range((len(columns))):
    plt.subplot(14,2,2*i+1)
    plt.hist(df[columns[i]])
    plt.title(columns[i]+"  Histogram")
```

```python
df1 = np.log1p(df[columns])
```

```python
features = df1.columns
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
sc = MinMaxScaler()
sc.fit(df1[features])
```

```python
features_sc = pd.DataFrame(sc.transform(df1[features]),columns  = df1.columns)
features_sc
```

```python
from sklearn.cluster import KMeans
```

```python
kmean = KMeans()
kmean.fit(features_sc)
```

```python
kmean.labels_
```

```python
sumofsquaredist = []
for i in range(1,11):
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(df)
    sumofsquaredist.append(kmeans.inertia_)
plt.plot(range(1,11),sumofsquaredist)
plt.title("Elbo method")
plt.xlabel("number of clusters")
plt.ylabel("inertia(sum of squared distance)")
plt.show()
```

```python
kmean = KMeans(n_clusters=3)
kmean.fit(features_sc)
kmean.labels_
```
